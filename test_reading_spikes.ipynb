```python
%matplotlib inline
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
from focal import *
```

```python
fname = "./mnist-db/spikes/train/0/"
fname += "image__idx_000001__lbl_00_"

# fname = "./mnist-db/spikes/t10k/7/"
# fname += "image__idx_000000__lbl_07_"

fname += ".npz"

img_w = 28
img_h = 28
img_shape = (img_h, img_w)
img_size = img_w * img_h
cmap='Greys_r'
```

```python
fcl = Focal()
```

```python
kernels = fcl.kernels.full_kernels
plt.figure(figsize=(12, 3))
for i in range(len(kernels)):
    ax = plt.subplot(1, 4, i + 1)
    ax.set_title(i)
    plt.imshow(kernels[i], cmap=cmap)
plt.show()
```

```python
data = np.load(fname)
print(data)
```

```python
print(data.keys())
```

```python
print(len(data['spikes']))
```

```python
n_neurons = len(data['spikes'])
n_neurons_per_type = n_neurons//4

print(n_neurons, n_neurons_per_type)
```

```python
spikes = data['spikes']
images = []
plt.figure(figsize=(12, 3))
for layer in range(4):
    start_idx = n_neurons_per_type * layer
    end_idx = start_idx + n_neurons_per_type
    
    img = np.zeros(img_shape)
    
    for pix, nid in range(start_idx, end_idx):
        row = (pix) // img_w
        col = (pix) % img_w
        
        if len(spikes[nid]):
            img[row, col] = 1
    
    images.append(img)
    ax = plt.subplot(1, 4, layer + 1)
    plt.imshow(img, cmap=cmap)

plt.show()
```

```python
spikes = data['spikes']
convs = []
for layer in range(4):
    start_idx = n_neurons_per_type * layer
    end_idx = start_idx + n_neurons_per_type
    
    img = np.zeros(img_shape)
    
    for nid in range(start_idx, end_idx):
        row = (nid - start_idx) // img_w
        col = (nid - start_idx) % img_w
        
        if len(spikes[nid]):
            img[row, col] = 1.0

    cnv = sp.signal.convolve2d(img, kernels[layer], mode='same')
    convs.append(cnv)
    plt.figure()
    ax = plt.subplot(1, 3, 1)
    plt.imshow(img, cmap=cmap)
    ax = plt.subplot(1, 3, 2)
    plt.imshow(kernels[layer], cmap=cmap)
    ax = plt.subplot(1, 3, 3)
    plt.imshow(cnv, cmap=cmap)
    
plt.show()
```

```python
rec = np.zeros(img_shape)
for c in convs:
    rec += c

plt.figure()
plt.imshow(rec, cmap=cmap)
plt.show()
```
