```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from focal import Focal, focal_to_spike, spike_trains_to_images_g
from scipy.signal import convolve2d
from scipy import misc
```

```python
face = misc.face() / 255.0
```

```python
plt.figure()
plt.imshow(face)
plt.show()
```

```python
RED, GREEN, BLUE = range(3)
yellow = (face[:, :, RED] * 0.299000 + face[:, :, GREEN] * 0.587000)
yellow2 = face.copy()
yellow2[:, :, BLUE] = 0

red = face[:, :, RED]
red2 = face.copy()
red2[:, :, GREEN] = 0
red2[:, :, BLUE] = 0

green = face[:, :, GREEN]
green2 = face.copy()
green2[:, :, RED] = 0
green2[:, :, BLUE] = 0

blue = face[:, :, BLUE]
blue2 = face.copy()
blue2[:, :, RED] = 0
blue2[:, :, GREEN] = 0

fw = 6
plt.figure(figsize=(2*fw, 4*fw))

plt.subplot(4, 2, 1)
plt.imshow(yellow, cmap='gray')
plt.subplot(4, 2, 2)
plt.imshow(yellow2)

plt.subplot(4, 2, 3)
plt.imshow(red, cmap='gray')
plt.subplot(4, 2, 4)
plt.imshow(red2)

plt.subplot(4, 2, 5)
plt.imshow(green, cmap='gray')
plt.subplot(4, 2, 6)
plt.imshow(green2)

plt.subplot(4, 2, 7)
plt.imshow(blue, cmap='gray')
plt.subplot(4, 2, 8)
plt.imshow(blue2)


plt.show()
```

```python
fig = plt.figure(figsize=(fw*3, fw*1))

ax = plt.subplot(1, 3, 1)
ax.set_title('blue - yellow')
im = plt.imshow(blue - yellow, cmap='gray')#, vmin=-0.5, vmax=0.5)
plt.colorbar(im)

ax = plt.subplot(1, 3, 2)
ax.set_title('red - green')
im = plt.imshow(red - green, cmap='gray')#, vmin=-1, vmax=1)
plt.colorbar(im)

ax = plt.subplot(1, 3, 3)
ax.set_title('achromatic')
im = plt.imshow(green * 0.587 + red * 0.299 + blue * 0.114, cmap='gray', vmin=0, vmax=1)
plt.colorbar(im)

plt.show()
```
