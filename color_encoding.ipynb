```python
# face = misc.face() / 255.0
face = misc.face() / 1.0
print(dir(misc))
```

```python
plt.figure()
plt.imshow(face.astype('int'))
plt.show()
```

```python
RED, GREEN, BLUE = range(3)
WRED, WGREEN = 0.7, 0.3
# WRED, WGREEN = 1.0, 0.0
# WRED, WGREEN = 0.299, 0.587
# WRED, WGREEN = 0.5, 0.5
# WRED, WGREEN = 0.5, 0.5
yellow = (face[:, :, RED] * WRED + face[:, :, GREEN] * WGREEN)
yellow2 = face.copy()
yellow2[:, :, BLUE] = 0
yellow2[:, :, RED] *= WRED
yellow2[:, :, GREEN] *= WGREEN

red = face[:, :, RED]
red2 = face.copy()
red2[:, :, GREEN] = 0
red2[:, :, BLUE] = 0

green = face[:, :, GREEN]
green2 = face.copy()
green2[:, :, RED] = 0
green2[:, :, BLUE] = 0

blue = face[:, :, BLUE]
blue2 = face.copy()
blue2[:, :, RED] = 0
blue2[:, :, GREEN] = 0

fw = 6
plt.figure(figsize=(2*fw, 4*fw))

plt.subplot(4, 2, 1)
plt.imshow(yellow, cmap='gray')
plt.subplot(4, 2, 2)
plt.imshow(yellow2.astype('int'))

plt.subplot(4, 2, 3)
plt.imshow(red, cmap='gray')
plt.subplot(4, 2, 4)
plt.imshow(red2.astype('int'))

plt.subplot(4, 2, 5)
plt.imshow(green, cmap='gray')
plt.subplot(4, 2, 6)
plt.imshow(green2.astype('int'))

plt.subplot(4, 2, 7)
plt.imshow(blue, cmap='gray')
plt.subplot(4, 2, 8)
plt.imshow(blue2.astype('int'))


plt.show()
```

```python
fig = plt.figure(figsize=(fw*3, fw*3))
by = blue - yellow
rg = red - green
achromatic = green * 0.587 + red * 0.299 + blue * 0.114

by_max = np.abs(by).max()
rg_max = np.abs(rg).max()
max_max = max(by_max, rg_max)

ax = plt.subplot(3, 3, 1)
ax.set_title('blue - yellow')
im = plt.imshow(by, cmap='plasma_r', vmin=-max_max, vmax=max_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 2)
ax.set_title('blue')
im = plt.imshow(by * (by > 0), cmap='gray', vmin=0.0, vmax=by_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 3)
ax.set_title('yellow')
im = plt.imshow(-by * (by < 0), cmap='gray', vmin=0.0, vmax=by_max)
plt.colorbar(im)


ax = plt.subplot(3, 3, 4)
ax.set_title('red - green')
im = plt.imshow(rg, cmap='RdYlGn_r', vmin=-max_max, vmax=max_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 5)
ax.set_title('red')
im = plt.imshow(rg * (rg > 0), cmap='gray', vmin=0.0, vmax=rg_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 6)
ax.set_title('green')
im = plt.imshow(-rg * (rg < 0), cmap='gray', vmin=0.0, vmax=rg_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 7)
ax.set_title('achromatic')
im = plt.imshow(achromatic, cmap='gray')
plt.colorbar(im)

plt.show()
```

```python
h = np.arange(5, blue.shape[0], 4)
w = np.arange(5, blue.shape[1], 4)

n_w = len(w)
n_h = len(h)
rows = np.repeat(h, n_w)
cols = np.tile(w, n_h)


```

```python
# blue_g = cv2.GaussianBlur(blue, (21, 21), 2)
# red_g = cv2.GaussianBlur(red, (21, 21), 2)
# green_g = cv2.GaussianBlur(green, (21, 21), 2)
# yellow_g = cv2.GaussianBlur(yellow, (21, 21), 2)
# achromatic_g = cv2.GaussianBlur(achromatic, (21, 21), 2)

blue_g = convolve(blue, g1d)
red_g = convolve(red, g1d)
green_g = convolve(green, g1d)
yellow_g = convolve(yellow, g1d)
achromatic_g = convolve(achromatic, g1d)



blue_gs = blue_g[rows, cols].reshape(n_h, n_w)
red_gs = red_g[rows, cols].reshape(n_h, n_w)
green_gs = green_g[rows, cols].reshape(n_h, n_w)
yellow_gs = yellow_g[rows, cols].reshape(n_h, n_w)
achromatic_gs = achromatic_g[rows, cols].reshape(n_h, n_w)
print(blue_gs.shape)
```

```python
# thr = 0.0#5

bmy0 = blue_gs - yellow_gs 
rmg0 = red_gs  - green_gs
# rmg0 = yellow_gs  - green_gs
# bmy0 = blue_gs * (blue_gs > thr) - yellow_gs * (yellow_gs > thr) 
# rmg0 = red_gs * (red_gs > thr) - green_gs * (green_gs > thr)

bmy = blue_gs * (bmy0 > 0) - yellow_gs * (bmy0 < 0)
rmg = red_gs * (rmg0 > 0) - green_gs * (rmg0 < 0)
# rmg = yellow_gs * (rmg0 > 0) - green_gs * (rmg0 < 0)
# bmy = bmy0
# rmg = rmg0
mx = max(np.max(np.abs(bmy)), np.max(np.abs(rmg)))

plt.figure(figsize=(3*fw, 2*fw))
ax = plt.subplot(2, 3, 1)
ax.set_title('blue')
im = plt.imshow(blue_gs, cmap='gray')#, vmin=0, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 2)
ax.set_title('yellow')
im = plt.imshow(yellow_gs, cmap='gray')#, vmin=0, vmax=1)
plt.colorbar(im)

bmy_max = np.max(np.abs(bmy))
ax = plt.subplot(2, 3, 3)
ax.set_title('blue - yellow')
im = plt.imshow(bmy, cmap='PuOr', vmin=-bmy_max, vmax=bmy_max)
plt.colorbar(im)


ax = plt.subplot(2, 3, 4)
ax.set_title('red')
im = plt.imshow(red_gs, cmap='gray')#, vmin=0, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 5)
ax.set_title('green')
im = plt.imshow(green_gs, cmap='gray')#, vmin=0, vmax=1)
plt.colorbar(im)

rmg_max = np.max(np.abs(rmg))
ax = plt.subplot(2, 3, 6)
ax.set_title('red - green')
im = plt.imshow(rmg, cmap='PiYG_r', vmin=-rmg_max, vmax=rmg_max)

plt.colorbar(im)


plt.show()
```

```python
thr = 0.05

rmg_out = rmg * (np.abs(rmg) > thr)
bmy_out = bmy * (np.abs(bmy) > thr)

a_out = bmy * (bmy > thr) 
b_out = -bmy * (bmy < -thr)
c_out = rmg * (rmg > thr)
d_out = -rmg * (rmg < -thr)
```

```python
#WRED, WGREEN
red_from_yellow = (b_out - d_out * WGREEN)/WRED
green_from_yellow = (b_out - d_out * WRED)/WGREEN
green_from_achromatic = (achromatic_gs - c_out * 0.299 - a_out * 0.114) / 0.587
height, width = a_out.shape
rec = np.zeros((height, width, 3))
rec[:, :, RED] = (b_out - d_out * WGREEN)/WRED
# rec[:, :, RED] = ((b_out + c_out) - d_out)
rec[:, :, RED] = c_out +  (c_out == 0) * red_from_yellow
# rec[:, :, GREEN] = d_out #+ green_from_yellow
rec[:, :, GREEN] = d_out + (d_out == 0) * green_from_achromatic
# rec[:, :, RED] = c_out + (c_out == 0) * red_from_yellow
# rec[:, :, GREEN] = d_out + (d_out == 0) * green_from_yellow


rec[:, :, BLUE] = a_out 

fig = plt.figure(figsize=(16, 7))
ax = plt.subplot(1, 3, 1)
im = plt.imshow(face.astype('int'))
cb = plt.colorbar(im)

ax = plt.subplot(1, 3, 2)
im = plt.imshow((rec*50).astype('int'))
cb = plt.colorbar(im)

ax = plt.subplot(1, 3, 3)
im = plt.imshow((15 * rec * achromatic_gs[:, :, np.newaxis]).astype('int'))
cb = plt.colorbar(im)

plt.show()
```

```python
big_h, big_w, channels = face.shape
big_rec = np.zeros((big_h, big_w, channels))
big_rec[rows, cols, RED] = rec[:, :, RED].reshape(len(rows))
big_rec[rows, cols, GREEN] = rec[:, :, GREEN].reshape(len(rows))
big_rec[rows, cols, BLUE] = rec[:, :, BLUE].reshape(len(rows))

fig = plt.figure(figsize=(10, 7))
ax = plt.subplot(1, 1, 1)
im = plt.imshow((big_rec*50).astype('int'))
cb = plt.colorbar(im)
plt.show()
```
