```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from focal import Focal, focal_to_spike, spike_trains_to_images_g
from scipy.signal import convolve2d
from scipy import misc
import cv2
```

```python
face = misc.face() / 255.0
```

```python
plt.figure()
plt.imshow(face)
plt.show()
```

```python
RED, GREEN, BLUE = range(3)
WRED, WGREEN = 0.7, 0.3
# yellow = (face[:, :, RED] * 0.299000 + face[:, :, GREEN] * 0.587000)
# yellow = (face[:, :, RED] * 0.5 + face[:, :, GREEN] * 0.5)
yellow = (face[:, :, RED] * WRED + face[:, :, GREEN] * WGREEN)
yellow2 = face.copy()
yellow2[:, :, BLUE] = 0
yellow2[:, :, RED] *= WRED
yellow2[:, :, GREEN] *= WGREEN

red = face[:, :, RED]
red2 = face.copy()
red2[:, :, GREEN] = 0
red2[:, :, BLUE] = 0

green = face[:, :, GREEN]
green2 = face.copy()
green2[:, :, RED] = 0
green2[:, :, BLUE] = 0

blue = face[:, :, BLUE]
blue2 = face.copy()
blue2[:, :, RED] = 0
blue2[:, :, GREEN] = 0

fw = 6
plt.figure(figsize=(2*fw, 4*fw))

plt.subplot(4, 2, 1)
plt.imshow(yellow, cmap='gray')
plt.subplot(4, 2, 2)
plt.imshow(yellow2)

plt.subplot(4, 2, 3)
plt.imshow(red, cmap='gray')
plt.subplot(4, 2, 4)
plt.imshow(red2)

plt.subplot(4, 2, 5)
plt.imshow(green, cmap='gray')
plt.subplot(4, 2, 6)
plt.imshow(green2)

plt.subplot(4, 2, 7)
plt.imshow(blue, cmap='gray')
plt.subplot(4, 2, 8)
plt.imshow(blue2)


plt.show()
```

```python
fig = plt.figure(figsize=(fw*3, fw*3))
by = blue - yellow
rg = yellow - green

by_max = np.abs(by).max()
rg_max = np.abs(rg).max()
max_max = max(by_max, rg_max)

ax = plt.subplot(3, 3, 1)
ax.set_title('blue - yellow')
im = plt.imshow(by, cmap='plasma_r', vmin=-max_max, vmax=max_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 2)
ax.set_title('blue')
im = plt.imshow(by * (by > 0), cmap='gray', vmin=0.0, vmax=by_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 3)
ax.set_title('yellow')
im = plt.imshow(-by * (by < 0), cmap='gray', vmin=0.0, vmax=by_max)
plt.colorbar(im)


ax = plt.subplot(3, 3, 4)
ax.set_title('red - green')
im = plt.imshow(rg, cmap='RdYlGn_r', vmin=-max_max, vmax=max_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 5)
ax.set_title('red')
im = plt.imshow(rg * (rg > 0), cmap='gray', vmin=0.0, vmax=rg_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 6)
ax.set_title('green')
im = plt.imshow(-rg * (rg < 0), cmap='gray', vmin=0.0, vmax=rg_max)
plt.colorbar(im)

ax = plt.subplot(3, 3, 7)
ax.set_title('achromatic')
im = plt.imshow(green * 0.587 + red * 0.299 + blue * 0.114, cmap='gray', vmin=0, vmax=1)
plt.colorbar(im)

plt.show()
```

```python
blue_g = cv2.GaussianBlur(blue, (21, 21), 2)
red_g = cv2.GaussianBlur(red, (21, 21), 2)
green_g = cv2.GaussianBlur(green, (21, 21), 2)
yellow_g = cv2.GaussianBlur(yellow, (21, 21), 2)

blue_gs = blue_g[10::4, 10::4]
red_gs = red_g[10::4, 10::4]
green_gs = green_g[10::4, 10::4]
yellow_gs = yellow_g[10::4, 10::4]
```

```python
bmy = blue_gs - yellow_gs
rmg = yellow_gs - green_gs


plt.figure(figsize=(3*fw, 2*fw))
ax = plt.subplot(2, 3, 1)
ax.set_title('blue')
im = plt.imshow(blue_gs, cmap='plasma_r', vmin=-1, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 2)
ax.set_title('yellow')
im = plt.imshow(yellow_gs, cmap='plasma', vmin=-1, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 3)
ax.set_title('blue - yellow')
im = plt.imshow(bmy * (np.abs(bmy) > 0.05), cmap='plasma_r', vmin=-1, vmax=1)
plt.colorbar(im)


ax = plt.subplot(2, 3, 4)
ax.set_title('red')
im = plt.imshow(red_gs, cmap='RdYlGn_r', vmin=-1, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 5)
ax.set_title('green')
im = plt.imshow(green_gs, cmap='RdYlGn', vmin=-1, vmax=1)
plt.colorbar(im)

ax = plt.subplot(2, 3, 6)
ax.set_title('red - green')
im = plt.imshow(rmg * (np.abs(rmg) > 0.05), cmap='RdYlGn_r', vmin=-1, vmax=1)
plt.colorbar(im)


plt.show()
```
