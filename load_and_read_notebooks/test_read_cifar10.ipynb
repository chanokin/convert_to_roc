```python
%matplotlib inline
import os
import sys
sys.path.insert(1, os.path.join(sys.path[0], '..'))

import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from focal import Focal, focal_to_spike, spike_trains_to_images_g
from scipy.signal import convolve2d

def todict(np_file):
    d = {}
    for k in list(np_file.keys()):
        try:
            d[k] = np_file[k].item()
        except:
            d[k] = np_file[k]
    return d

def ssa_to_img(ssa, scale, img_shape, power_exp=-1.0, start_time=0.0, end_time=np.inf):
    max_places = 4 * np.prod(img_shape)
    max_t = np.max([np.max(ts) for ts in ssa if len(ts)])
    t2p = max_places / max_t
    h, w = img_shape
    n_per_scale = h * w
    start = scale * n_per_scale
    end = start + n_per_scale
    img = np.zeros(img_shape)
    for pix, neuron in enumerate(range(start, end)):
        times = np.asarray(ssa[neuron])
        whr = np.where(np.logical_and(start_time <= times, times < end_time))[0]
        n_spikes = len(whr)
        if n_spikes > 0:
            places = (times[whr] * t2p) + 1
            img[pix//w, pix%w] = np.sum(places**(power_exp))

    
    return img
        
```

```python
dir_name = "../cifar_output/class_000000"
files = glob.glob(os.path.join(dir_name, '*'))
print(len(files))

spikes = []
bmg_spikes = []
rmg_spikes = []
fw = 3.
for f in [files[2]]:
    data = todict(np.load(f, allow_pickle=True))
    print(data.keys())
    kernels = data['kernels']
    shape = data['grayscale_image'].shape

    spikes[:] = data['focal_spikes']
    images = spike_trains_to_images_g(spikes, data['grayscale_image'])

    conv_images = [
        convolve2d(images[k], kernels[k], mode='same')
        for k in images
    ]
        
    s = np.sum(conv_images, axis=0)
    if s.min() < 0:
        s -= s.min()
    s /= s.max()

    spikes[:] = data['focal_spikes']
    images = spike_trains_to_images_g(spikes, data['grayscale_image'])

    conv_images = [
        convolve2d(images[k], kernels[k], mode='same')
        for k in images
    ]
        
    gray_s = np.sum(conv_images, axis=0)
#     gray_s[gray_s < 0] = 0
    if gray_s.min() < 0:
        gray_s -= gray_s.min()
    gray_s /= gray_s.max()


    # -----------------------------------------------------------  #

    bmg_spikes[:] = data['bmg_spikes']
    images = spike_trains_to_images_g(bmg_spikes, data['bmg_image'])

    conv_images = [
        convolve2d(images[k], kernels[k], mode='same')
        for k in images
    ]
        
    bmg_s = np.sum(conv_images, axis=0)
#     bmg_s[bmg_s < 0] = 0
    if bmg_s.min() < 0:
        bmg_s -= bmg_s.min()
    bmg_s /= bmg_s.max()


    # -----------------------------------------------------------  #

    rmg_spikes[:] = data['rmg_spikes']
    images = spike_trains_to_images_g(rmg_spikes, data['rmg_image'])

    conv_images = [
        convolve2d(images[k], kernels[k], mode='same')
        for k in images
    ]
        
    rmg_s = np.sum(conv_images, axis=0)
    if rmg_s.min() < 0:
        rmg_s -= rmg_s.min()
    rmg_s /= rmg_s.max()

    # -----------------------------------------------------------  #

    
    plt.figure(figsize=(10, 3))
    plt.suptitle('L : reconstruction')
    ax = plt.subplot(1, 3, 1)
    ax.set_title('recons')
    im = plt.imshow(gray_s, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)
    
    ax = plt.subplot(1, 3, 2)
    ax.set_title('gray')
    im = plt.imshow(data['grayscale_image'], cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)

    ax = plt.subplot(1, 3, 3)
    ax.set_title('color')
    im = plt.imshow(data['color_image'], vmin=0, vmax=1)
    plt.colorbar(im)
    
    # -----------------------------------------------------------  #

    plt.figure(figsize=(10, 3))
    plt.suptitle('B - L : reconstruction')
    ax = plt.subplot(1, 3, 1)
    ax.set_title('recons')
    im = plt.imshow(bmg_s, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)
    
    ax = plt.subplot(1, 3, 2)
    ax.set_title('gray')
    im = plt.imshow(data['bmg_image'], cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)

    ax = plt.subplot(1, 3, 3)
    ax.set_title('color')
    im = plt.imshow(data['color_image'], vmin=0, vmax=1)
    plt.colorbar(im)

    # -----------------------------------------------------------  #

    plt.figure(figsize=(10, 3))
    plt.suptitle('R - L : reconstruction')
    ax = plt.subplot(1, 3, 1)
    ax.set_title('recons')
    im = plt.imshow(rmg_s, cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)
    
    ax = plt.subplot(1, 3, 2)
    ax.set_title('gray')
    im = plt.imshow(data['rmg_image'], cmap='gray', vmin=0, vmax=1)
    plt.colorbar(im)

    ax = plt.subplot(1, 3, 3)
    ax.set_title('color')
    im = plt.imshow(data['color_image'], vmin=0, vmax=1)
    plt.colorbar(im)



    plt.show()
    

    
# ============================================================ #    
# ============================================================ #
# ============================================================ #
    
    
    
wred, wgreen, wblue = data['wred'], data['wgreen'], data['wblue']
rmg_s -= 0.5
bmg_s -= 0.5
# gray_s -(rmg_s * wred + bmg_s * wblue) / wgreen
green = gray_s - ( (rmg_s * wred + bmg_s * wblue) / wgreen )
# green = ( gray_s - (rmg_s * wred + bmg_s * wblue) ) 
red = gray_s + rmg_s
blue = gray_s + bmg_s

rec = np.zeros((gray_s.shape[0], gray_s.shape[1], 3))
rec[:, :, 0] = red
rec[:, :, 1] = green
rec[:, :, 2] = blue


plt.figure(figsize=(10, 3))
plt.suptitle('R - L : reconstruction')
ax = plt.subplot(1, 3, 1)
ax.set_title('gray')
im = plt.imshow(gray_s, cmap='gray', vmin=0, vmax=1)
plt.colorbar(im)

ax = plt.subplot(1, 3, 2)
ax.set_title('color recons')
im = plt.imshow(rec)#, vmin=0, vmax=1)
plt.colorbar(im)

ax = plt.subplot(1, 3, 3)
ax.set_title('color')
im = plt.imshow(data['color_image'], vmin=0, vmax=1)
plt.colorbar(im)
```
